```python
#Convert one datatype into another from Object to Numeric.
df['Pitch Speed [mph]'] = pd.to_numeric(df['Pitch Speed [mph]'], errors ='coerce' )
df['Pitch Spin [rpm]'] = pd.to_numeric(df['Pitch Spin [rpm]'], errors ='coerce' )
df['Pitch Release Height [ft]'] = pd.to_numeric(df['Pitch Release Height [ft]'], errors ='coerce' )
df['Pitch Extension [ft]'] = pd.to_numeric(df['Pitch Extension [ft]'], errors ='coerce' )
```
